```python
df["rating_category"] = "Unknown"

df.loc[df["Rating"] <= 2.5, "rating_category"] = "Poor"
df.loc[(df["Rating"] > 2.5) & (df["Rating"] <= 3.5), "rating_category"] = "Average"
df.loc[(df["Rating"] > 3.5) & (df["Rating"] <= 4.5), "rating_category"] = "Good"
df.loc[df["Rating"] > 4.5, "rating_category"] = "Excellent"
```

```python

# 1. List all restaurants in a specific locality.
locality = "SM Megamall| Ortigas| Mandaluyong City"

result = df[df["Locality"] == locality][["RestaurantName","Rating"]]

print(result)

```

```python
# 2. Show the name and city of restaurants with a rating lower than 3.

df[df['Rating']<3][['RestaurantName','City']]
```

```python
# 3. Count the number of restaurants that do not offer table booking.
a = df[df["Has_Table_booking"]=='No']["RestaurantID"].count()
print(a)
```

```python
# 4. Display restaurants that have an average cost for two less than $50.
currency_map = {
    "Botswana Pula(P)": "BWP",
    "Brazilian Real(R$)": "BRL",
    "Dollar($)": "USD",
    "Emirati Diram(AED)": "AED",
    "Indian Rupees(Rs.)": "INR",
    "Indonesian Rupiah(IDR)": "IDR",
    "NewZealand($)": "NZD",
    "Pounds(?)": "GBP",
    "Qatari Rial(QR)": "QAR",
    "Rand(R)": "ZAR",
    "Sri Lankan Rupee(LKR)": "LKR",
    "Turkish Lira(TL)": "TRY"
}
df["Currency_Code"] = df["Currency"].map(currency_map)
exchange_to_usd = {
    "USD": 1.00,
    "BWP": 0.079,
    "BRL": 0.19,
    "AED": 0.27,
    "INR": 0.012,
    "IDR": 0.000067,
    "NZD": 0.60,
    "GBP": 1.26,
    "QAR": 0.27,
    "ZAR": 0.054,
    "LKR": 0.0031,
    "TRY": 0.052
}
df["Cost_in_USD"] = df["Average_Cost_for_two"] * df["Currency_Code"].map(exchange_to_usd)
df[df["Cost_in_USD"]<50]["RestaurantName"]
```

```python
# 5. Find restaurants with more than 100 votes.
df[df["Votes"]>100]['RestaurantName']
```

```python
# 6. List the names of restaurants offering both table booking and online delivery.

df[(df["Has_Table_booking"] == "Yes") & (df["Has_Online_delivery"] == 'Yes')]["RestaurantName"]
```

```python
# 7. Show the cuisines offered by restaurants in 'Paris'.

paris_cuisines = df.loc[df["City"] == "Paris", "Cuisines"]
print(paris_cuisines)
```

```python

# 8. Count the number of restaurants in each price range.

df['Price_range'].value_counts()
```

```python
# 9. Display the top 10 restaurants by rating in a specific country.
CountryCode = "1"

result = (
    df[df["CountryCode"] == CountryCode]
    [["RestaurantName", "Rating"]]
    .sort_values(by="Rating", ascending=False)
    .head(10)
)

print(result)
```

```python
# 10. List all restaurants with 'Cafe' in their cuisine type.
result = df[df["Cuisines"].str.contains("Cafe", case=False, na=False)]["RestaurantName"]
print(result)
```

```python
avg_votes_locality = df.groupby("Locality")["Votes"].mean().reset_index()
print(avg_votes_locality)
```

```python
online_delivery_city = df[df["Has_Online_delivery"] == "Yes"].groupby("City")["RestaurantName"].count().reset_index()
online_delivery_city.rename(columns={"RestaurantName": "Online_Delivery_Count"}, inplace=True)
print(online_delivery_city)
```

```python
table_booking_city = df[df["Has_Table_booking"] == "Yes"].groupby("City")["RestaurantName"].count()
table_booking_city = table_booking_city.sort_values(ascending=False).reset_index()
print(table_booking_city.head())
```

```python
chinese_rating = df[df["Cuisines"].str.contains("Chinese", na=False)]
avg_chinese_rating = chinese_rating["Rating"].mean()
print(avg_chinese_rating)
```

```python
most_common_price_london = df[df["City"] == "London"]["Price_range"].mode()[0]
print(most_common_price_london)
```

```python
# Split cuisines by comma and explode
df_cuisines = df.copy()
df_cuisines["CuisineList"] = df_cuisines["Cuisines"].str.split(",")
df_exploded = df_cuisines.explode("CuisineList")

cuisine_count_country = df_exploded.groupby("CountryCode")["CuisineList"].nunique().reset_index()
cuisine_count_country.rename(columns={"CuisineList": "Total_Cuisines"}, inplace=True)
print(cuisine_count_country)
```

```python
min_cost_city = df.loc[df.groupby("City")["Average_Cost_for_two"].idxmin()][["City", "RestaurantName", "Average_Cost_for_two"]]
print(min_cost_city)
```

```python
top_localities = df["Locality"].value_counts().head(5).reset_index()
top_localities.columns = ["Locality", "Restaurant_Count"]
print(top_localities)
```

```python
delivery_percentage = df.groupby("City")["Is_delivering_now"].value_counts(normalize=True).mul(100).rename("percentage").reset_index()

# Filter only delivering now
delivery_percentage = delivery_percentage[delivery_percentage["Is_delivering_now"] == "Yes"]
print(delivery_percentage[["City", "percentage"]])
```
